```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
psms = pd.read_table('../data/peak_width/peak_width_PSMs.txt')
```

```python
psms['FWHM'] = psms['FWHM'].replace(0, pd.NA)
psms = psms[~psms['FWHM'].isna()]
psms = psms[psms['FWHM']<60]
psms = psms.drop_duplicates(subset=['Annotated Sequence', 'Charge'])
```

```python
fig, ax = plt.subplots(figsize=(4, 2))
psms['FWHM'].hist(bins=15, ax=ax)
ax.set_xlabel('FWHM [s]')
ax.set_title(f'Peak width of {len(psms)} identified precursors')
plt.savefig('../results/peak_width/peak_width.pdf', transparent=True, bbox_inches='tight')
```
